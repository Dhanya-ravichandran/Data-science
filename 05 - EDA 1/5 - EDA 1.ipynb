```python
# Fill missing categorical values with mode
for col in df.select_dtypes(include=[np.number]).columns:
    df[col].fillna(df[col].median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

Based on the `df.info()` output, all columns are already of type `float64`, which is suitable for numerical analysis. No data type conversion is required.

```python
# Plot boxplots for each numeric column
plt.figure(figsize=(15,8))
sns.boxplot(data=df.select_dtypes(include=[np.number]))
plt.xticks(rotation=45)
plt.title("Boxplots for Outlier Detection")
plt.show()
```

```python
# --- Outlier Treatment using IQR ---
numeric_cols = df.select_dtypes(include=[np.number]).columns

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[col] = np.where(df[col] < lower_bound, lower_bound, df[col])
    df[col] = np.where(df[col] > upper_bound, upper_bound, df[col])

print("\nData cleaning completed!")
```

```python
# Plot boxplots for each numeric column
plt.figure(figsize=(15,8))
sns.boxplot(data=df.select_dtypes(include=[np.number]))
plt.xticks(rotation=45)
plt.title("Boxplots for Outlier Detection")
plt.show()
```

**Standard Deviation and IQR**: These measures provide insights into the spread or dispersion of the data. A larger standard deviation or IQR indicates greater variability in the data for that column.

**Kurtosis**: This measure indicates the "tailedness" of the distribution. A kurtosis value close to 0 suggests a distribution similar to a normal distribution. Positive kurtosis indicates heavier tails (more outliers), while negative kurtosis indicates lighter tails. Most features show negative kurtosis, suggesting lighter tails than a normal distribution.

```python
df.hist(bins=20, figsize=(15, 12), edgecolor='black')
plt.suptitle("Histograms of Numeric Features", fontsize=16)
plt.show()
```

```python
bins = [0, 120, 160, 250]
labels = ['Low', 'Normal', 'High']
df['LB_category'] = pd.cut(df['LB'], bins=bins, labels=labels)

df['LB_category'].value_counts().plot.bar()
plt.title("Baseline FHR by Category")
plt.ylabel("Number of Cases")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Features")
plt.show()
```

```python
# --- 5. Violin Plots (distribution by class) ---
if "NSP" in df.columns:
    plt.figure(figsize=(15,8))
    for i, col in enumerate(df.drop(columns="NSP").columns[:6], 1):  # first 6 features
        plt.subplot(2, 3, i)
        sns.violinplot(x="NSP", y=col, data=df, palette="Set2")
        plt.title(f"{col} by NSP")
    plt.tight_layout()
    plt.show()
```
